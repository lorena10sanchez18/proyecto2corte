##  **Descripcion **

Antes de analizar una imagen debemos estar seguros de que la calidad de esta es suficiente. Esto es relativamente sencillo en aplicaciones industriales ya que el operador controla y fija la iluminación y todos los parámetros de la cámara mientras que la diversidad de casos que se puede capturar es limitado. No ocurre  asi en aplicaciones en exteriores donde la iluminación no es controlable y la diversidad de apariencia  notable.

Sin embargo, sí que se puede seguir controlando la cámara y a veces la óptica, ya que existen modelos donde se puede cambiar el tiempo que la cámara esté activa, abrir o cerrar la óptica o Modificar el enfoque. Una opción entonces es adquirir varias imágenes con diferentes parámetros en las capturas y elegir la mejor para analizar.

### Funciones y librerias

**Detencion de bordes canny**: 
El operador de detencion de bordes de canny fue desarrollado por el catedratico de la universidad de berkeley en 1986 y se basa en un algoritmo de multiples fases para detectar un amplio rango de bordes. Es sin duda el operador mas utilizado en la detencion de bordes.

**Calchist Histograma**: 
Un histograma es una representacion grafica de la distribucion de los niveles de grises en una imagen, utilizando el metodo de ecualizacion de histogramas podemos obtener una distribución uniforme del histograma de la imagen, este método mejora el contraste y brillo de una imagen que sea demasiado oscura o brillante, obteniendo un balance optimo de los pixeles blancos y negros.

**Sobel**: 
El operador Sobel es utilizado en procesamiento de imágenes, especialmente en algoritmos de detección de bordes. Técnicamente es un operador diferencial discreto que calcula una aproximación al gradiente de la función de intensidad de una imagen. Para cada punto de la imagen a procesar, el resultado del operador Sobel es tanto el vector gradiente correspondiente como la norma de este vector.

```python
#Codigo de Funciones para histogramas , opacidad y promedio de imágenes

import cv2
import numpy as np
from matplotlib import pyplot as plt

def Difuminacion():
    
    img = cv2.imread('Imagen1.jpg')
    blur = cv2.blur(img,(10,10))

    plt.subplot(121),plt.imshow(img),plt.title('Original')
    plt.xticks([]), plt.yticks([])
    plt.subplot(122),plt.imshow(blur),plt.title('Difuminada')
    plt.xticks([]), plt.yticks([])
    
def Histogramas():
    
    img = cv2.imread('Imagen2=1.jpg')
    cv2.imshow('Imagen1=2.jpg', img)
    color = ('b','g','r')
    
    for i, c in enumerate(color):
        hist = cv2.calcHist([img], [i], None, [256], [0, 256])
        plt.plot(hist, color = c)
        plt.xlim([0,256])
    plt.show()
    plt.imshow(img),plt.title("Imagen  1 que parece a la 2")
    
def Promedio():
    img = cv2.imread('Imagen2.jpg')
#Crea el kernel
    kernel = np.ones((7,7),np.float32)/25
 
    #Filtra la imagen utilizando el kernel anterior
    dst = cv2.filter2D(img,-1,kernel)
    blur = cv2.blur(img,(1,1))
    plt.subplot(121),plt.imshow(img),plt.title('Imagen Original tren')
    plt.xticks([]), plt.yticks([])
    plt.subplot(122),plt.imshow(dst),plt.title(' Imagen del tren promediadad')
    plt.xticks([]), plt.yticks([])
    
    plt.show()
    
Promedio()
Histogramas()
Difuminacion()

plt.show()
```

```python
#Codigo histograma color

import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('imagen2=1.jpg')
cv2.imshow('imagen1=2.jpg', img)

color = ('b','g','r')

for i, c in enumerate(color):
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(hist, color = c)
    plt.xlim([0,256])

plt.show()
plt.imshow(img),plt.title("Imagen  1 que parece a la 2")
cv2.destroyAllWindows()
```

```python
#histograma escala de grises
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('mujer.png', cv2.IMREAD_GRAYSCALE)
cv2.imshow('mujer.png', img)

hist = cv2.calcHist([img], [0], None, [256], [0, 256])
plt.plot(hist, color='gray' )

plt.xlabel('intensidad de iluminacion')
plt.ylabel('cantidad de pixeles')
plt.show()
plt.imshow(img),plt.title("Imagen 6 Tren")
cv2.destroyAllWindows()
```
